```python
data['x0'] = data['x0'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x1'] = data['x1'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x2'] = data['x2'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x3'] = data['x3'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x4'] = data['x4'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x5'] = data['x5'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x6'] = data['x6'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x7'] = data['x7'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x8'] = data['x8'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x9'] = data['x9'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x10'] = data['x10'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x11'] = data['x11'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x12'] = data['x12'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['x13'] = data['x13'].astype(str).str.replace('[(,),$,%]', '',regex=True)
data['y'] = data['y'].astype(str).str.replace('[(,),$,%]', '',regex=True)
```

```python
data.rename(columns={'x0': '0','x1': '1','x2': '2','x3': '3','x4': '4','x5': '5','x6': '6','x7': '7','x8': '8','x9': '9','x10': '10','x11': '11','x12': '12','x13': '13'}, inplace='True')
```

```python
data = data.astype(float)
data.describe()
```

```python
data.corr()
```

```python
x = data[['10','9']]
y = data['y']
```

```python
data['10'].fillna((data['10'].mean()), inplace=True)
data['9'].fillna((data['9'].mean()), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

stand_data = StandardScaler()
X_train = stand_data.fit_transform(X_train)
X_test = stand_data.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=11)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error

mse_knn = mean_squared_error(y_true=y_train, y_pred=y_pred_knn)
mse_knn

# I was able to get a result when I ran my code previously and it was about 0.0079.  I don't know what changed between the instances and solutions found did not fix issue.
```
