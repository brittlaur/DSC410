Label encoding all categorical data to make it easier to use in model

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df['encoded_Agency'] = le.fit_transform(df['Agency'])
df
```

```python
df['encoded_Agency_Type'] = le.fit_transform(df['Agency Type'])
df['encoded_Distribution'] = le.fit_transform(df['Distribution Channel'])
df['encoded_Product'] = le.fit_transform(df['Product Name'])
df['encoded_Claim'] = le.fit_transform(df['Claim'])
df
```

```python
encoded_features = ['encoded_Agency', 'encoded_Agency_Type', 'encoded_Distribution', 'encoded_Product']
numerical_features = ['Duration', 'Net Sales', 'Age', 'Commision (in value)']

X = df[numerical_features + encoded_features]
y = df['encoded_Claim'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 101)
```

```python
#standardize numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train[numerical_features + encoded_features])
X_train.loc[:, numerical_features + encoded_features] = scaler.transform(X_train[numerical_features + encoded_features])
```

Decision tree was used because a logistic model and KNN didn't work in my mind due to columns given wouldn't affect why a person would make a claim. So it would make it too difficult for the model to use the training data.  

```python
from sklearn.tree import DecisionTreeClassifier
class_tree = DecisionTreeClassifier(max_depth=3)
class_tree.fit(X_train, y_train)
```

```python
import six
from six import StringIO
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(decision_tree=class_tree,
                out_file=dot_data,
                filled=True,
                rounded=True,
                feature_names=X_train.columns,
                class_names = ['0','1'],
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
class_tree = DecisionTreeClassifier(max_depth=6, min_samples_split=50)
class_tree.fit(X_train, y_train)
y_pred_class_tree = class_tree.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score

accuracy_class_tree = accuracy_score(y_true=y_train,
                                     y_pred=y_pred_class_tree)
accuracy_class_tree
```

```python
pd.Series(data=class_tree.feature_importances_,
          index=X_train.columns).sort_values(ascending=False).round(3)
```

Created random forest classifier to further test the accuracy of the training data.

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=99,
                            max_features=5,
                            max_depth=4,
                            min_samples_split=100,
                            random_state=85)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_train)
```

```python
accuracy_rf = accuracy_score(y_true=y_train, y_pred=y_pred_rf)
accuracy_rf
```

```python
pd.Series(data=rf.feature_importances_,
          index=X_train.columns).sort_values(ascending=False).round(3)
```

```python
#Test the model

y_pred_null = np.zeros_like(y_test)
accuracy_score(y_true=y_test, y_pred = y_pred_null)
```

Validation of Model

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix

def CM(y_true, y_pred):
    M = confusion_matrix(y_true, y_pred)
    out = pd.DataFrame(M, index=['Obs No', 'Obs Yes'],
                                columns=['Pred No', 'Pred Yes'])
    return out                                 
```

```python
threshold = 0.5
y_pred_prob = rf.predict_proba(X_test)[:,1]
y_pred = (y_pred_prob > threshold).astype(int)
```

```python
CM(y_test, y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Precision:  {:0.1f}%, Recall: {:.1f}%'.format(100*precision, 100*recall))
```

It makes sense that the precision and recall are low because ultimately the categories used to predict if the person would make a claim would not make a difference.  The correlation between claims and any category is low, which would make it harder to make predictions based on that.  It also does not help that the amount of people that did make claims is only 1.5% of the data, giving the model no reason to not predict a no.

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.kdeplot(y_pred_prob[y_test==0], shade=True, color='red',
label='No', ax=ax)
sns.kdeplot(y_pred_prob[y_test==1], shade=True, color='green',
label='Yes', ax=ax)
ax.set_title('Distribution of predicted probabilities', fontsize=16)
ax.legend()
plt.grid();
```
